```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris, load_wine
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
```

```python
data = load_iris()
X = data.data
y = data.target
target_names = data.target_names

print("Original Data Shape:", X.shape)
```

```python
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)
```

```python
normalizer = MinMaxScaler()
X_normalized = normalizer.fit_transform(X_standardized)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_normalized)
```

```python
print("\nExplained Variance Ratio:")
for i, var in enumerate(pca.explained_variance_ratio_):
    print(f"Principal Component {i+1}: {var:.4f}")

print("Cumulative Explained Variance:",
      np.cumsum(pca.explained_variance_ratio_))
```

```python
plt.figure()
for i in np.unique(y):
    plt.scatter(X_pca[y == i, 0],
                X_pca[y == i, 1],
                label=target_names[i])

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA - 2D Visualization (Iris Dataset)")
plt.legend()
plt.show()
```

```python
pca_full = PCA()
X_pca_full = pca_full.fit_transform(X_normalized)

plt.figure()
plt.plot(np.cumsum(pca_full.explained_variance_ratio_), marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Scree Plot (Cumulative Explained Variance)")
plt.grid(True)
plt.show()
```

```python
plt.figure()

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Original Data")

plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.title("PCA Transformed Data")

plt.tight_layout()
plt.show()
```
